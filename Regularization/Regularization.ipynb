### Regularization (Ridge L2, Lasso L1)
Purpose of regularization is to achieve low Bias and Variance

To reduce multicollinearity

```python
from warnings import filterwarnings
filterwarnings('ignore')
```

```python
import pandas as pd
df = pd.read_csv('Cars93.csv', na_values=['', 'NA'], keep_default_na=False)
df.head()
```

### Step 2: Perform basic data quality checks

### Step 3: Seperate X and Y(Weight)

```python
x = df.drop(columns=['id','Weight'])
y = df[['Weight']]
```

```python
x.head()
```

```python
y.head()
```

### Step 4: Apply preprocessing on X

```python
cat = list(x.columns[x.dtypes=='object'])
con = list(x.columns[x.dtypes!='object'])
```

```python
cat
```

```python
con
```

```python
from sklearn.pipeline import Pipeline
from sklearn.impute import SimpleImputer
from sklearn.preprocessing import StandardScaler, OneHotEncoder
from sklearn.compose import ColumnTransformer
```

```python
num_pipe = Pipeline(steps=[('impute', SimpleImputer(strategy='mean')),
                    ('scaler', StandardScaler())])
```

```python
cat_pipe = Pipeline(steps=[('impute', SimpleImputer(strategy='most_frequent')),
                           ('ohe', OneHotEncoder(handle_unknown='ignore', sparse_output=False))])
```

```python
pre = ColumnTransformer([('num', num_pipe, con),
                         ('cat', cat_pipe, cat)]).set_output(transform='pandas')
```

```python
x_pre = pre.fit_transform(x)
x_pre.head()
```

```python
pre
```

### Step 5: Apply train test split

```python
from sklearn.model_selection import train_test_split
xtrain, xtest, ytrain, ytest = train_test_split(x_pre, y, test_size=0.2, random_state=60)
```

```python
xtrain.head()
```

```python
ytrain.head()
```

```python
xtest.head()
```

```python
ytest.head()
```

```python
xtrain.shape
```

```python
xtest.shape
```

### Step 6: Start building the models

```python
from sklearn.linear_model import LinearRegression
model1 = LinearRegression()
model1.fit(xtrain,ytrain)
```

```python
model1.score(xtest, ytest)
```

```python
from sklearn.linear_model import Ridge
model2 = Ridge(alpha=1)
model2.fit(xtrain ,ytrain)
```

```python
model2.score(xtest, ytest)
```

```python
from sklearn.linear_model import Lasso
model3 = Lasso(alpha=1)
model3.fit(xtrain, ytrain)
```

```python
model3.score(xtest, ytest)
```

### Step 7 : hyperparameter tuning - GridSearchCV

alpha - Hyperparameter

What is optimal value of alpha for ridge and lasso for a given dataset

```python
import numpy as np
params = {'alpha': np.arange(start=0.1, stop=100, step=0.1)}
```

```python
params
```

```python
# Tuning for ridge
from sklearn.model_selection import GridSearchCV
rr = Ridge()
gscv1 = GridSearchCV(estimator=rr, param_grid=params, cv=5, scoring='neg_mean_squared_error')
gscv1.fit(xtrain, ytrain)
```

```python
gscv1.best_params_
```

```python
gscv1.best_score_
```

```python
best_ridge = gscv1.best_estimator_
best_ridge
```

```python
best_ridge.score(xtrain, ytrain)
```

```python
best_ridge.score(xtest, ytest)
```

```python
# Tuning for Lasso
ls = Lasso()
gscv2 = GridSearchCV(estimator=ls, param_grid=params, cv=5, scoring='neg_mean_squared_error')
gscv2.fit(xtrain, ytrain)
```

```python
gscv2.best_params_
```

```python
gscv2.best_score_
```

```python
best_lasso = gscv2.best_estimator_
best_lasso
```

```python
best_lasso.score(xtrain, ytrain)
```
